```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Define the categorical columns
categorical_columns = ['Country', 'Product', 'Import_Export', 'Category', 'Port', 'Customs_Code', 
                       'Shipping_Method', 'Supplier', 'Customer', 'Payment_Terms']

# Function to compute Cramér's V for categorical correlation
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()  # Total observations
    r, k = confusion_matrix.shape  # Number of rows and columns
    return np.sqrt(chi2 / (n * (min(r, k) - 1)))

# Dictionary to hold statistics for each column
descriptive_stats = {}

# Initialize tables for each statistic
count_table = pd.DataFrame(index=categorical_columns, columns=['Count'])
frequency_table = {}
proportion_table = {}
min_table = pd.DataFrame(index=categorical_columns, columns=['Minimum'])
max_table = pd.DataFrame(index=categorical_columns, columns=['Maximum'])
mode_table = pd.DataFrame(index=categorical_columns, columns=['Mode'])
rank_table = {}
correlation_table = pd.DataFrame(index=categorical_columns, columns=categorical_columns)

# Calculate Count, Frequency, Proportion, Minimum, Maximum, Mode, Rank, and Correlation
for col in categorical_columns:
    counts = pdataset[col].value_counts()  # Get frequency counts of each category
    total_count = pdataset[col].count()  # Non-null count
    mode_val = pdataset[col].mode()[0]  # Most frequent category
    rank_vals = counts.rank(ascending=False, method='min')  # Rank based on frequency
    
    # Fill the respective statistics tables
    count_table.loc[col] = total_count
    frequency_table[col] = counts
    proportion_table[col] = (counts / total_count)
    min_table.loc[col] = pdataset[col].min()
    max_table.loc[col] = pdataset[col].max()
    mode_table.loc[col] = mode_val
    rank_table[col] = rank_vals

# Calculate Correlation (Cramér's V) for categorical columns
for col1 in categorical_columns:
    for col2 in categorical_columns:
        if col1 != col2:
            correlation_table.loc[col1, col2] = cramers_v(pdataset[col1], pdataset[col2])
        else:
            correlation_table.loc[col1, col2] = 1.0  # Correlation with itself is 1

# Convert the frequency and proportion dictionaries to DataFrames
frequency_df = pd.DataFrame(frequency_table)
proportion_df = pd.DataFrame(proportion_table)
rank_df = pd.DataFrame(rank_table)
#------------------------------------------------------------------------------------------
```

# Z - TEST

1. QUANTITY - checking proportion of quantity between 0 to 2000

```python
Pro_Quantity = 0 # Sample proportion
for i in range(len(non_cat_stats['Quantity'])):
    n = non_cat_stats['Quantity'].iloc[i]
    if (0 <= n <= 2000):
        Pro_Quantity += 1
    
Size_Quantity = len(non_cat_stats['Quantity'])

print('Sample proporation: ', Pro_Quantity)
print('Sample size: ', Size_Quantity)

hypothesize_proportion = 500  # Hypothesized quantity proportion

print('hypothesized proportion', hypothesize_proportion)

standard_error = (hypothesize_proportion * (1 - hypothesize_proportion) / Size_Quantity) ** 0.5

z_stats = (Pro_Quantity - hypothesize_proportion) / standard_error

p_value = 2 * (1 - stats.norm.cdf(abs(z_stats)))

print('Z-stats: ', z_stats)
print('p_value: ', p_value)

significance_level = 0.05

if p_value < significance_level:
    print('hypothesized proportion is significantly different from the proportion quantity')
else:
    print('hypothesized proportion is not significantly different from the proportion quantity')
```

# 1. Shapiro-Wilk Test

```python
def shapiro_test(sample, name):
    stat, p_value = stats.shapiro(sample)
    print("Shapiro-wilk Test for: ", name)
    print("Test stats: ", stat, " and p-value: ", p_value)

    significance_level = 0.05
    if p_value > significance_level:
        print("Fail to reject (sample is consistent with normality)")
    else:
        print("Reject (sample does not appear to be normally distributed)")


shapiro_test(non_cat_df['Quantity'], "Quantity")
print("\n")
shapiro_test(non_cat_df['Weight'], "Weight")
print("\n")
shapiro_test(non_cat_df['Value'], "Value")
```
